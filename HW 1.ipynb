```python
#This dataset attributes first names to genders, giving counts and probabilities. It combines open-source government data from the US, UK, Canada, and Australia.
import pandas as pd
data=pd.read_csv("name_gender_dataset.csv")
```

```python
name=set(data["Name"])
gender=set(data["Gender"])
count=set(data["Count"])
probability=set(data["Probability"])
```

```python
#Q1
#超過1000000人取的名字
over1m_names=data[data["Count"]>=1000000]["Name"]
over1m_names
```

```python
#Q2
#超過150000人且A開頭的名字
over150t_Astart_names=data[(data["Count"]>=150000)&(data["Name"].str.startswith("A"))]["Name"]
over150t_Astart_names
```

```python
#Q3
#機率介於0.002-0.0025的名字
probability_around002to0025_names=data[(data["Probability"]>=0.002)&(data["Probability"]<=0.0025)]["Name"]
probability_around002to0025_names
```

```python
#Q4
#男生女生名字哪個比較多
num_males=len(data[data["Gender"]=="M"])
num_females=len(data[data["Gender"]=="F"])

if num_males>num_females:
    print("男生名字比較多")
elif num_females>num_males:
    print("女生名字比較多")
else:
    print("男女名字一樣多")
    
print("男性名字數為",num_males)
print("女性名字數為",num_females)
print("兩者相差名字數為",abs(num_males-num_females))
```

```python
#Q5
#最多人取的100個名字
top_100_names=data.sort_values("Count",ascending=False).head(100)["Name"]
top_100_names
```

```python
#Q6
#前100個最多人取的名字中，上下兩名平均人數差距
top_100_names_num=data.sort_values("Count",ascending=False).head(100)
average=float(0)
for i in range(0,100,2):
    count_up=top_100_names_num.iloc[i]["Count"]
    count_down=top_100_names_num.iloc[i+1]["Count"]
    average+=(count_up-count_down)
    
average=average/99
average
```

```python
#Q7
#T開頭y結尾的名字
StartWithT=set(data[data["Name"].str.startswith("T")]["Name"])

EndsWithY=set(data[data["Name"].str.endswith("y")]["Name"])

StartWithT&EndsWithY
```

```python
#Q8
#男生女生一樣的名字有哪些
male_names=set(data[data["Gender"]==("M")]["Name"])
female_names=set(data[data["Gender"]==("F")]["Name"])
male_names&female_names
```

```python
#Q9
#男生B開頭的名字跟女生B開頭的名字哪個比較多
num_males_B=len(data[(data["Gender"]=="M")&(data["Name"].str.startswith("B"))]["Name"])
num_females_B=len(data[(data["Gender"]=="F")&(data["Name"].str.startswith("B"))]["Name"])
if num_males_B>num_females_B:
    print("B開頭的男生名字比較多")
elif num_females_B>num_males_B:
    print("B開頭的女生名字比較多")
else:
    print("B開頭的男女名字一樣多")
    
print("B開頭的男性名字數為",num_males_B)
print("B開頭的女性名字數為",num_females_B)
print("B開頭的兩者相差數目為",abs(num_males_B-num_females_B))
```

```python
#Q10
#前100多的男生名字是G開頭的機率
```

```python
probability_100Male_Gstart=float(0)
probability_100Male=float(0)

malenames_num=data[data["Gender"]==("M")]
top_100_malenames_num=malenames_num.sort_values("Count",ascending=False).head(100)
for i in range(0,100):
    probability_100Male+=top_100_malenames_num.iloc[i]["Probability"]
    if top_100_malenames_num.iloc[i]["Name"].startswith("G"):
        probability_100Male_Gstart+=top_100_malenames_num.iloc[i]["Probability"]
    
probability_100Male_Gstart/probability_100Male
```
