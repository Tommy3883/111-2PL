```python
#匯入檔案(spotify kpop2021,2022最多收聽數藝人及團體)
import pandas as pd
data2021=pd.read_csv("2021.csv")
data2021
```

```python
data2022=pd.read_csv("2022.csv")
data2022
```

```python
#列出子集合s1
s1=set(data2021["artist"])
s1
```

```python
#列出子集合s1
s2=set(data2022["artist"])
s2
```

```python
#交集
#2021與2022皆有上榜的團體或歌手
s1&s2
print("2021與2022皆有上榜的團體或歌手數量:",len(s1&s2))
s1&s2
```

```python
#聯集
#2021或2022有上榜的團體或歌手
s1|s2
print("2021或2022有上榜的團體或歌手數量:",len(s1|s2))
s1|s2
```

```python
#差集2021
#2021有上榜但2022沒有上榜的團體或歌手
s1-(s1&s2)
print("2021有上榜但2022沒有上榜的團體或歌手:",len(s1-(s1&s2)))
s1-(s1&s2)
```

```python
#差集2022
#2022有上榜但2021沒有上榜的團體或歌手
s2-(s1&s2)
print("2022有上榜但2021沒有上榜的團體或歌手:",len(s2-(s1&s2)))
s2-(s1&s2)
```

```python
#對稱差集
#只有上榜2021或2022其中一年的團體或歌手
(s1|s2)-(s1&s2)
print("只有上榜2021或2022其中一年的團體或歌手",len((s1|s2)-(s1&s2)))
(s1|s2)-(s1&s2)
```
