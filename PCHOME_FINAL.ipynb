```python
import requests
from bs4 import BeautifulSoup
import time
```

```python
# get web data
num = 0
data = []
page = 0

while num < 300:
# (check data length)
    page = page + 1
    # print(page)
    # f:format, 將資料格式化

    # import url
    url = f'https://ecshweb.pchome.com.tw/search/v3.3/all/results?q=%E7%AD%86%E9%9B%BB&page={page}&sort=rnk/dc'

    a = requests.get(url)
    result = a.json()
    
    data.append(result['prods'])
    num += 20

    time.sleep(0.5)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(data)
```

```python
def flatten_list(_2d_list):
    flat_list = []
    # Iterate through the outer list
    for element in _2d_list:
        if type(element) is list:
            # If the element is of type list, iterate through the sublist
            for item in element:
                flat_list.append(item)
        else:
            flat_list.append(element)
    return flat_list
```

```python
df1 = flatten_list(data)
```

```python
df2 = pd.DataFrame(df1)
df2
```
