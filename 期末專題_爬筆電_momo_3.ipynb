```python
import requests 
import json
import csv
import time
from time import sleep
import pandas as pd
import re
from bs4 import BeautifulSoup
from urllib.parse import urljoin

# 讀取 laptop_all_links.csv 檔案，並取得網址列表
links = []
with open('laptop_all_links.csv', 'r', newline='', encoding='utf-8') as csvfile:
    reader = csv.reader(csvfile)
    next(reader)  # 跳過表頭
    for row in reader:
        url = row[0]
        links.append(url)

# 開啟 laptop_info.csv 檔案並寫入配備資訊
with open('laptop_info.csv', 'w', newline='', encoding='utf-8-sig') as csvfile:
    writer = csv.writer(csvfile)

    # 寫入表頭
    writer.writerow(['網址', '圖片網址', '價錢', '品牌', '作業系統', '記憶體', '容量', '尺寸', '處理器'])

    # 逐個網址進行爬取
    for url in links:
        # 發起請求並獲取網頁內容
        response = requests.get(url)
        time.sleep(5)
        html = response.content

        # 使用 Beautiful Soup 解析網頁內容
        soup = BeautifulSoup(html, 'html.parser')

        # 爬取配備資訊
        image_url = soup.select_one('#goodsimg > a > div > div.zoomPup > img')['src']
        price = soup.select_one('#productForm > div.prdwarp > div.prdnoteArea.jsCartFloat > ul.prdPrice > li.special > span').text.strip()
        brand = soup.select_one('#attributesTable > tbody > tr:nth-child(1) > td > ul > li > a').text.strip()
        operating_system = soup.select_one('#attributesTable > tbody > tr:nth-child(2) > td > ul > li').text.strip()
        memory = soup.select_one('#attributesTable > tbody > tr:nth-child(3) > th').text.strip()
        storage = soup.select_one('#attributesTable > tbody > tr:nth-child(4) > th').text.strip()
        size = soup.select_one('#attributesTable > tbody > tr:nth-child(6) > th').text.strip()
        processor = soup.select_one('#attributesTable > tbody > tr:nth-child(10) > th').text.strip()

        # 寫入一行資料
        writer.writerow([url, image_url, price, brand, operating_system, memory, storage, size, processor])
```
