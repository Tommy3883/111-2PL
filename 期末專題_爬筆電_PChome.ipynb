```python
url = 'https://24h.pchome.com.tw/region/DHAF'

# 發送 GET 請求取得網頁內容
response = requests.get(url)

# 使用 BeautifulSoup 解析 HTML
soup = BeautifulSoup(response.text, 'html.parser')
print(soup)

# 尋找商品連結
product_links = []
container = soup.select('#Block12Container50')[0]
items = container.select('dd:nth-child(1) a.prod_img')
for item in items:
    product_link = 'https://24h.pchome.com.tw' + item['href']
    product_links.append(product_link)

# 輸出結果
for product_link in product_links:
    print(product_link)
```

```python
import requests
import csv
from bs4 import BeautifulSoup

url = 'https://24h.pchome.com.tw/prod/DHAS1U-1900FUROL?q=/S/DHAFNK'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# 找到內容元素
content_element = soup.select_one('#root > div > main > div > div > div > section:nth-child(2) > div > div > section.u-mt24 > div > div.col-10 > div:nth-child(4) > div > div > p')

# 獲取內容
content = content_element.get_text(strip=True)

# 寫入 CSV 檔案
with open('pchome_item_data.csv', 'w', newline='', encoding='utf-8-sig') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['Content'])
    writer.writerow([content])
```
