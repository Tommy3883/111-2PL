```python
import requests 
import json
import csv
import time
from time import sleep
import pandas as pd
import re
from bs4 import BeautifulSoup

# 設定 User-Agent

headers = {
    'cookie': 'over18=1', 
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.82 Safari/537.36'
}

# 發送 GET 請求取得文章內容
url = "https://f7best.com/brand-laptop-series/"
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, "html.parser")

if response.status_code == 200:
    print(f'請求成功：{response.status_code}')
else:
    print(f'請求失敗：{response.status_code}')
```

```python
# 建立 CSV 檔案並寫入標頭
with open('laptop_links.csv', 'w', newline='') as csv_file:
    writer = csv.writer(csv_file)
    writer.writerow(['Brand', 'Series', 'Link'])

    # 從所有的 <a> 標籤中選取包含 "/product/" 的網址
    for link in soup.find_all('a', href=True):
        if '/product/' in link['href']:
            # 取得品牌和系列
            brand, series = link['href'].split('/')[4:6]

            # 寫入 CSV 檔案
            writer.writerow([brand, series, link['href']])
```

```python
# 讀取 csv 檔案中的所有網址
urls = []
with open('laptop_links.csv', 'r', newline='') as csv_file:
    reader = csv.DictReader(csv_file)
    for row in reader:
        urls.append(row['Link'])

# 儲存資訊的列表
data = []

# 爬取每個網址中的資訊
for url in urls:
    res = requests.get(url)
    soup = BeautifulSoup(res.text, 'html.parser')
    time.sleep(0.5)

    # 處理器
    processor_elem = soup.find('div', text='處理器：')
    print(processor_elem)
    if processor_elem:
        cpu = processor_elem.find_next_sibling('div').text.strip()
        print(cpu)
    else:
        cpu = None

    # 記憶體
    memory_tag = soup.find('div', text='記憶體：')
    if memory_tag:
        memory = processor_elem.find_next_sibling('div').text.strip()
    else:
        memory = None

    # 硬碟
    storage = None
    storage_tag = soup.find('div', text='硬碟：')
    if storage_tag:
        storage = storage_tag.find_next_sibling('div').text.strip()

    # 獨立顯卡
    gpu = None
    gpu_tag = soup.find('div', text='獨立顯卡：')
    if gpu_tag:
        gpu = gpu_tag.find_next_sibling('div').text.strip()

    # 螢幕
    screen = None
    screen_tag = soup.find('div', text='螢幕：')
    if screen_tag:
        screen = screen_tag.find_next_sibling('div').text.strip()

    # 無線網路
    wireless = None
    wireless_tag = soup.find('div', text='無線網路：')
    if wireless_tag:
        wireless = wireless_tag.find_next_sibling('div').text.strip()

    # 介面
    interface = None
    interface_tag = soup.find('div', text='介面：')
    if interface_tag:
        interface = interface_tag.find_next_sibling('div').text.strip()

    # 電池
    battery = None
    battery_tag = soup.find('div', text='電池：')
    if battery_tag:
        battery = battery_tag.find_next_sibling('div').text.strip()

    # 重量
    weight = None
    weight_tag = soup.find('div', text='重量：')
    if weight_tag:
        weight = weight_tag.find_next_sibling('div').text.strip()

    # 作業系統
    os = None
    os_tag = soup.find('div', text='作業系統：')
    if os_tag:
        os = os_tag.find_next_sibling('div').text.strip()

    # 保固
    warranty = None
    warranty_tag = soup.find('div', text='保固：')
    if warranty_tag:
        warranty = warranty_tag.find_next_sibling('div').text.strip()

    # 將每筆資料存到 data 列表中
    data.append([cpu, memory, storage, gpu, screen, wireless, interface, battery, weight, os, warranty])

# 將 data 寫入 csv 檔案
with open('laptop_info.csv', 'w', newline='') as csv_file:
    writer = csv.writer(csv_file)
    writer.writerow(['處理器', '記憶體', '硬碟', '獨立顯卡', '螢幕', '無線網路', '介面', '電池', '重量', '作業系統', '保固'])
    for row in data:
        writer.writerow(row)
print("finish")
```
